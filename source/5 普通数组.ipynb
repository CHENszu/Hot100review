## 5.1 最大子数组和

输入：nums = [-2,1,-3,4,-1,2,1,-5,4]  
输出：6  
解释：连续子数组 [4,-1,2,1] 的和最大，为 6 。

```python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        # 初始化：当前连续子数组的最大和（以上一个元素结尾）
        current_max = nums[0]
        # 初始化：全局最大和（最终答案）
        global_max = nums[0]
        
        # 从第二个元素开始遍历数组
        for num in nums[1:]:
            # 核心公式：更新当前最优解
            current_max = max(num, current_max + num)
            # 更新全局最优解
            global_max = max(global_max, current_max)
        
        return global_max
```

## 5.2 合并区间

输入：intervals = [[1,3],[2,6],[8,10],[15,18]]  
输出：[[1,6],[8,10],[15,18]]  
解释：区间 [1,3] 和 [2,6] 重叠, 将它们合并为 [1,6].

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        if not intervals:  # 处理空数组边界情况
            return []
        # 1. 按区间左边界升序排序
        intervals.sort()
        res = [intervals[0]]  # 初始化结果数组，放入第一个区间
        # 2. 遍历剩余区间，逐个合并
        for curr in intervals[1:]:
            last = res[-1]
            if curr[0] <= last[1]:
                # 重叠则合并：更新最后一个区间的右边界
                last[1] = max(last[1], curr[1])
            else:
                # 不重叠则追加
                res.append(curr)
        return res
```

## 5.3 轮转数组


输入: nums = [1,2,3,4,5,6,7], k = 3  
输出: [5,6,7,1,2,3,4]

```python
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        k = k % n  # 有效轮转步数
        nums[:] = nums[-k:] + nums[:-k]  # 原地修改数组，必须写 nums[:]
```

## 5.4 除了自身以外数组的乘积

输入: nums = [1,2,3,4]  
输出: [24,12,8,6]

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1] * n
        
        # 第一步：计算左侧乘积，存入res
        for i in range(1, n):
            res[i] = res[i-1] * nums[i-1]
        
        # 第二步：用变量维护右侧乘积，倒序遍历更新res
        right_num = 1
        for i in range(n-1, -1, -1):
            res[i] *= right_num
            right_num *= nums[i]
        
        return res
```
